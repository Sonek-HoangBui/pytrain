```python
from enum import Enum

class Gender(Enum):
    MALE = 1
    FEMALE = 2
    OTHER = 3

class CanBo:
    _name: str
    _age: int
    _gender: Gender
    _addr: str

    def __init__(self, name, age, gender, addr=""):
        self._name = name
        self._age = age
        self._gender = gender 
        self._addr = addr
        
    def inputInfo(self):
        self._name = str(input("Input name: "))
        self._age = int(input("Input age: "))
        self._gender = str(input("Inputgender: "))
        if self._gender == Gender.FEMALE:
            print("Gender: ", self.gender)
        elif self.gender == Gender.MALE:
            print("Gender: ", self.gender)
        else:
            print("Gender: Khác")
            
        self.addr = str(input("Input addr: "))
        
    def showInfo(self):
        print("name: ", self._name)
        print("age: ", self._age)
        print("gender: ", self._gender)
        print("addr: ", self._addr)


canbo = CanBo('Nguyệt', 16, Gender.FEMALE)
canbo.showInfo()
```

```python
MIN_LEVEL = 0
MAX_LEVEL = 10

class CongNhan(CanBo):
    #__Level = ''
    _level: int
    def __init__(self, name, age, gender, addr, level = MIN_LEVEL):
        super().__init__(name, age, gender, addr)
        self._level = level
        
    def inputInfo(self):
        super().inputInfo()
        self._level = int(input("Input level: "))
        if self._level >= MIN_LEVEL and self._level <= MAX_LEVEL:
            print(self._level)
        else:
            print("Cấp bậc chỉ từ 1 đến 10")
            self._level = int(input("Input level: "))
        
    def showInfo(self):
        super().showInfo()
        print("level: ", self._level)
        
        
class KySu(CanBo):
    #__Nganh  = ''
    _nganh: str
    def __init__(self, name, age, gender, addr, NganhDaoTao):
        super().__init__(name, age, gender, addr)
        self._nganh = NganhDaoTao
    # def setNganh(self):
    #     return 0
    def inputInfo(self):
        super().inputInfo()
        self._nganh = input("Input NganhDaoTao: ")
        
    def showInfo(self):
        super().showInfo()
        print("NganhDaoTao: ", self._nganh)
        
        
class NhanVien(CanBo):
    #__Task = ''
    _task: str
    def __init__(self, name, age, gender, addr, task):
        super().__init__(name, age, gender, addr)
        self._task = task
    # def setTask(self):
    #     return 0
    
    def inputInfo(self):
        super().inputInfo()
        self._task = input("Input Task: ")
        
    def showInfo(self):
        super().showInfo()
        print("Task: ", self._task)

cn = CongNhan("Huyền", 16, Gender.MALE, "", 10)
nv = NhanVien("Nguyên", 16, Gender.FEMALE, "", "Báo cáo")
ks = KySu("Ánh", 16, Gender.FEMALE, "", "IT")

cn.showInfo()
nv.showInfo()
ks.showInfo()
```

```python
class QuanLyCB:
    _listCanBo: list[CanBo]
    
    def AddCB(self): # ok
        name = input("Nhap ten can bo: ")
        age = input("Nhap tuoi can bo: ")
        gender = input("Nhap gioi tinh can bo: ")
        addr = input("Nhap dia chi can bo: ")
        cb = CanBo(name, age, gender, addr)
        self._listCanBo.append(cb)
        
    def AddCB1(self, name, age, gender, addr): # ok
        cb = CanBo(name, age, gender, addr)
        self._listCanBo.append(cb)

    def FindName(self, key): # ok
        listCB = []
        if (self._listCanBo.__len__() > 0):
            for cb in self._listCanBo:
                if (key.upper() in cb._name.upper()):
                    listCB.append(cb)
        return listCB
            
    def showCB(self): # ok
        for cb in self._listCanBo:
            cb.showInfo()

    def delete(self, name):
        newListCB = []
        for cb in self._listCanBo:
            if name.casefold() == cb._name.casefold():
                print("delete ", name.casefold(), cb._name.casefold(), " ===")
            else:
                newListCB.append(cb)

        self._listCanBo = newListCB


qlCB = QuanLyCB()
qlCB.AddCB1("Huyền", 16, Gender.MALE, "Hà Nội")
qlCB.AddCB1("Nguyệt", 16, Gender.MALE, "Hà Nội")
qlCB.AddCB1("Ánh", 16, Gender.MALE, "Hà Nội")

print("=== show all after add ===")
qlCB.showCB() # show all
qlCB.delete("Nguyệt")
print("=== show after delete ===")
qlCB.showCB() # show 2 nguoi
result = qlCB.FindName("Nguyệt")
print("=== show result search after delete Nguyet ===")
for cb in result: # ko co ai
    cb.showInfo() 

print("=== show result search after delete Nguyet -- Anh ===")
result1 = qlCB.FindName("Ánh")
for cb in result1:
    cb.showInfo()
```
